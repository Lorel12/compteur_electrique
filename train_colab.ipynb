```
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from ultralytics import YOLO
from torchvision import transforms
import torch.nn as nn
import pytesseract
import re

yolo_model = YOLO('runs/detect/train2/weights/best.pt')  # <-- adapte le chemin si besoin

class DigitCNN(nn.Module):
    def __init__(self):
        super(DigitCNN, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        return self.fc(self.conv(x))

# === Load trained digit CNN model
digit_model = DigitCNN()
digit_model.load_state_dict(torch.load("/content/Prelevement/Prelevement/cnn_digit_segment.pt", map_location='cpu'))
digit_model.eval()

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

def preprocess_digit(roi):
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    resized = cv2.resize(gray, (28, 28), interpolation=cv2.INTER_AREA)
    _, thresh = cv2.threshold(resized, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return transform(thresh).unsqueeze(0)

def predict_digit(roi):
    tensor = preprocess_digit(roi)
    with torch.no_grad():
        output = digit_model(tensor)
        return torch.argmax(output, dim=1).item()

def extract_info(img_path):
    img = cv2.imread(img_path)
    results = yolo_model(img)[0]

    index_pred = ''
    compteur_pred = ''

    for box in results.boxes.data:
        cls = int(box[5].item())
        label = results.names[cls]
        x1, y1, x2, y2 = map(int, box[:4])
        roi = img[y1:y2, x1:x2]

        if label == 'index':

                height = y2 - y1
                roi_top = roi[0:int(height * 0.9), :]

                # PrÃ©-traitement (niveaux de gris + seuillage
                gray = cv2.cvtColor(roi_top, cv2.COLOR_BGR2GRAY)
                _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

                # 7-segment display
                result = reader.readtext(thresh, allowlist='0123456789')

                detected_text = ""
                roi_boxes_display = cv2.cvtColor(thresh, cv2.COLOR_GRAY2BGR) # For visualization

                for (bbox, text, prob) in result:
                    #[top-left, top-right, bottom-right, bottom-left]

                    x_min = int(min(point[0] for point in bbox))
                    y_min = int(min(point[1] for point in bbox))
                    x_max = int(max(point[0] for point in bbox))
                    y_max = int(max(point[1] for point in bbox))

                    clean_text = re.sub(r'[^0-9]', '', text) # non-digit
                    if clean_text:
                        detected_text += clean_text
                        cv2.rectangle(roi_boxes_display, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
                        cv2.putText(roi_boxes_display, clean_text, (x_min, y_min - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)


                plt.figure(figsize=(10,8))
                plt.imshow(cv2.cvtColor(roi_boxes_display, cv2.COLOR_BGR2RGB))
                plt.axis('off')
                plt.title(f"Chiffres dÃ©tectÃ©s dans l'index ({detected_text})")
                plt.show()

                #return detected_text.strip()

        elif label == 'compteur':
            # OCR pour extraire numÃ©ro de compteur
            roi_gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
            roi_gray = cv2.bilateralFilter(roi_gray, 11, 17, 17)
            text = pytesseract.image_to_string(roi_gray, config='--psm 6')
            text = text.replace('O', '0').replace('I', '1').replace('|', '1')
            numbers = re.findall(r'\d{6,12}', text)
            if numbers:
                compteur_pred = max(numbers, key=len)

    return detected_text, compteur_pred #index_pred

#
img_path = "/content/MINE/MINE/images/val/k.jpg"
index, compteur = extract_info(img_path)

print(" Index dÃ©tectÃ© :", index)
print(" NumÃ©ro de compteur :", compteur if compteur else "Non dÃ©tectÃ©")
```

```
import torch
import torch.nn as nn
from torchvision import transforms
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Le modÃ¨le CNN utilisÃ©
class DigitCNN(nn.Module):
    def __init__(self):
        super(DigitCNN, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        return self.fc(self.conv(x))

# Charger le modÃ¨le
model = DigitCNN()
model.load_state_dict(torch.load("/content/cnn_digit.pt", map_location="cpu"))
model.eval()

# Transformations identiques Ã  l'entraÃ®nement
transform = transforms.Compose([
    transforms.Resize((28, 28)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
```

```
def extract_digits(image_path, model):
    # Chargement image
    original = cv2.imread(image_path)
    gray = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)

    x, y, w, h = 100, 50, 200, 70  # Ã  ajuster selon ton image
    index_roi = gray[y:y+h, x:x+w]
    _, thresh_roi = cv2.threshold(index_roi, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    contours, _ = cv2.findContours(thresh_roi, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    # ensuite traitement uniquement sur ces contours


    # Extraire et trier les bounding boxes (gauche Ã  droite)
    boxes = [cv2.boundingRect(c) for c in contours]
    boxes = sorted(boxes, key=lambda b: b[0])  # Tri par position X

    index_digits = []
    for x, y, w, h in boxes:
        if w > 10 and h > 20:  # Filtrage des petits objets
            roi = gray[y:y+h, x:x+w]
            roi = cv2.resize(roi, (28, 28))
            pil_img = Image.fromarray(roi).convert("L")

            # Transformation et prÃ©diction
            input_tensor = transform(pil_img).unsqueeze(0)
            output = model(input_tensor)
            pred = output.argmax(dim=1).item()
            index_digits.append(str(pred))

            # Optionnel : visualiser la boÃ®te
            cv2.rectangle(original, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Affichage final
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    plt.title("Index dÃ©tectÃ© : " + "".join(index_digits))
    plt.axis("off")
    plt.show()

    return "".join(index_digits)
```

```
image_path = "/content/Prelevement/Prelevement/images/val/k.jpg"  # mets ici ton image
index = extract_digits(image_path, model)
print("âœ… Index extrait :", index)
```
